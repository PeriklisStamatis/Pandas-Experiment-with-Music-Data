# Music Dataset

The following is some basic Pandas DataFrame manipulation of a csv dataset I made manually of some of my favourite music albums

```python
#importing the neccessary libraries
#Pandas for handling/manipulating data
#Matplotlib for plotting and visualisation of said data

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#opening music.csv in reading mode in order to read the data
#and turn it into a DataFrame
#Opening this way ensures that the file will be closed when we 
#no longer need it

with open('Music.csv','r') as f:
    frame = pd.read_csv('Music.csv')
```

```python
#checking that the DataFrame was read properly
frame
```

```python
#Alternatively we can just check the head of the dataframe
#Since the dataframe isn't that big for now, we can display the whole thing.

frame.head()
```

```python
#displaying basic information about the DataFrame
frame.info()
```

```python
#filling the empty values in Genre 2 with the values form Genre 1
frame['Genre 2'] = frame['Genre 2'].fillna(frame['Genre 1'])
```

```python
#displaying the tail of the frame to check if the values are filled
frame.tail()
```

```python
#displaying certain basic statistical information about the dataframe (mean, standard deviation etc)
frame.describe()
```

Let's say we want to find the average rating of a Heavy Metal Album (according to Genre 1)

```python
#First we make a list of boolean values, where True means that Genre 1 for the 
#corresponding album is Heavy Metal, and False means it isn't. We need this 
#list in order to isolate the Heavy Metal Albums.
heavy_list = frame['Genre 1'] == 'Heavy Metal'
```

```python
#We use the list as an index in order to get rows that correspond to True
#which are the Heavy Metal Album (According to Genre 1)
#We store these as a new dataframe for easier handling and manipulation
#in the future
heavy_frame=frame[heavy_list]
heavy_frame
```

```python
#now we apply the mean to the Rating column of the new dataframe, to find the mean rating of 
#the Heavy Metal albums in the list
heavy_frame['Rating'].apply('mean')
```

The number is 9.25 as of now which shows my preference with the genre

Let's say we want to plot the rating of these albums

```python
#We make a list containing the ratings
heavy_ratings = list(heavy_frame['Rating'])

#And another list containing the corresponding names
heavy_names = list(heavy_frame['Album Name'])

plt.barh(heavy_names, heavy_ratings)
```
